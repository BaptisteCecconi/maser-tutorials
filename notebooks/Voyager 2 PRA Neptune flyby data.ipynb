```python
%matplotlib inline
import matplotlib.pyplot as plt 
import matplotlib.mlab as mlab
#from maser.data import cdpp
from maser.data.pds.ppi.voyager import pra
import datetime
import numpy
import os
root_data_path = os.path.join('/Users/baptiste/PycharmProjects/maser4py', 'maser', 'data', 'tests', 'data', 'pds') 
```

```python
file = os.path.join(root_data_path, 'VG2-N-PRA-3-RDR-LOWBAND-6SEC-V1', 'VG2_NEP_PRA_6SEC.LBL')
o = pra.PDSPPIVoyagerPRARDRLowBand6SecDataFromLabel(file, load_data=True) 
```

```python
t = o.time
f = o.frequency
raw_data = numpy.zeros((len(f),len(t)), dtype=numpy.int32)
for ii in range(len(o.object['TABLE'].data['DATE'])*8):
    raw_data[:,ii] = o.get_single_sweep(ii).raw_sweep[1:]
```

```python
plt.figure(figsize=(20,5))
plt.pcolormesh(t,f,raw_data, vmin=2000, vmax=3000, cmap='PuBu_r')
plt.ylim((0,500))
plt.xlabel(' SCET')
plt.xlim((datetime.datetime(1989,8,28,0,0),datetime.datetime(1989,8,28,6,0)))
plt.ylabel('Frequency (kHz)')
plt.title('Voyager 2 PRA at Neptune\nPower Spectral Density')
```
